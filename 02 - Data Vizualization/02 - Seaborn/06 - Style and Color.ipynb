## Style and Color

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(data = tips, x = 'sex')
```

```python
sns.set_style(style = 'whitegrid')
sns.countplot(data = tips, x = 'sex')
```

<code>style</code> can get one of:
<ul>
    <li>darkgrid</li>
    <li>whitegrid</li>
    <li>dark</li>
    <li>white</li>
    <li>ticks</li>
</ul>

```python
sns.set_style(style = 'darkgrid')
sns.countplot(data = tips, x = 'sex')
```

```python
sns.set_style(style = 'ticks')
sns.countplot(data = tips, x = 'sex')
```

### Spine Removal

```python
sns.countplot(data = tips, x = 'sex')
sns.despine()
```

```python
sns.countplot(data = tips, x = 'sex')
sns.despine(left = True)
```

### Size and Aspect

You can use matplotlib's <code>plt.figure(figsize=(width,height))</code> to change the size of most seaborn plots.

You can control the size and aspect ratio of most seaborn grid plots by passing in parameters: <code>size</code>, and <code>aspect</code>. For example:

```python
plt.figure(figsize=(12, 3)) # Non Grid Plot
sns.countplot(data = tips, x = 'sex')
```

```python
sns.lmplot(data = tips, hue = 'sex', palette = 'seismic',
           x = 'total_bill', y = 'tip',
          size = 2, aspect = 4)
```

### Scale and Context

The <code>set_context()</code> allows you to override default parameters:

```python
sns.set_context(context = 'poster')
sns.countplot(data = tips, x = 'sex')
```

```python
sns.set_context(context = 'poster', font_scale = 2)
sns.countplot(data = tips, x = 'sex')
```

<code>context</code> can get one of:
<ul>
    <li>paper</li>
    <li>notebook</li>
    <li>talk</li>
    <li>poster</li>
</ul>
