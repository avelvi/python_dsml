## Titles, labels, and legends

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.linspace(0, 5, 11)
y = x ** 2
```

### Titles

A title can be added to each axis instance in a figure. To set the title, use the <code>set_title</code> method in the axes instance:

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('Title')
```

### Labels

Similarly, with the methods <code>set_xlabel</code> and <code>set_ylabel</code>, we can set the labels of the X and Y axes:

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('Title')


ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
```

### Legends

You can use the <code>label="label text"</code> keyword argument when plots or other objects are added to the figure, and then using the <code>legend</code> method without arguments to add the legend to the figure:

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2)
ax.plot(x, x**3)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2)
ax.plot(x, x**3)
ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label="X Squared")
ax.plot(x, x**3, label="X Cubed")
ax.legend(loc = 0) # 'best' location
```

The <code>legend</code> function takes an optional keyword argument <code>loc</code> that can be used to specify where in the figure the legend is to be drawn. The allowed values of <code>loc</code> are numerical codes for the various places the legend can be drawn. See the documentation page for details. Some of the most common <code>loc</code> values are:
<li>1 - upper right corner</li>
<li>2 - upper left corner</li>
<li>3 - lower left corner</li>
<li>4 - lower right corner</li>

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label="X Squared")
ax.plot(x, x**3, label="X Cubed")
ax.legend(loc = 1)
```

```python
ax.legend(loc = 2)
fig
```

```python
ax.legend(loc = 3)
fig
```

```python
ax.legend(loc = 4)
fig
```

Also <code>loc</code> can take a tuple of (a, b) coordinats:

```python
ax.legend(loc = (0.2, 0.2))
fig
```
