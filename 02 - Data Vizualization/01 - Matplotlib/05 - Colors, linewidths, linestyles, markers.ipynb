## Colors, linewidths, linestyles, markers

### Colors

#### Colors with MatLab like syntax

With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where <code>'b'</code> means blue, <code>'g'</code> means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, <code>'b.-'</code> means a blue line with dots:

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, 'b')
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

#### Colors with the color= parameter

We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the color and alpha keyword arguments. Alpha indicates opacity.

```python
fig, ax = plt.subplots()

ax.plot(x, x + 1, color = "blue", alpha = 0.5, label = 'x + 1') # half-transparant
ax.plot(x, x + 2, color = "#8B008B", label = 'x + 2')        # RGB hex code
ax.plot(x, x + 3, color = "#FF8C00", label = 'x + 3')        # RGB hex code 
ax.legend()
```

### Linewidth

To change the line width, we can use the <code>linewidth</code> or <code>lw</code> keyword argument:

```python
fig, ax = plt.subplots()

ax.plot(x, x + 1, color = "blue", alpha = 0.5, label = 'x + 1', linewidth = 2)
ax.plot(x, x + 2, color = "#8B008B", label = 'x + 2')
ax.plot(x, x + 3, color = "#FF8C00", label = 'x + 3', lw = 0.5)
ax.legend()
```

### Linestyle

The line style can be selected using the <code>linestyle</code> or <code>ls</code> keyword arguments:

```python
fig, ax = plt.subplots()

# possible linestype options -, , -., :, steps
ax.plot(x, x + 1, color = "blue", label = 'x + 1', linestyle = '-')
ax.plot(x, x + 2, color = "#8B008B", label = 'x + 2', ls = '-.')
ax.plot(x, x + 3, color = "#FF8C00", label = 'x + 3', ls = ':')
ax.plot(x, x + 4, color = "red", label = 'x + 4', ls = 'steps')
ax.legend()
```

### Markers

```python
fig, ax = plt.subplots()

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x + 1, color = "blue", label = 'x + 1', marker = 'o')
ax.plot(x, x + 2, color = "#8B008B", label = 'x + 2', marker = '*')
ax.plot(x, x + 3, color = "#FF8C00", label = 'x + 3', marker = 's')
ax.plot(x, x + 4, color = "red", label = 'x + 4', marker = '3')
ax.plot(x, x + 5, color = "green", label = 'x + 5', marker = '+')
ax.legend()
```

With <code>marker</code> you can use <code>markersize</code>, <code>markerfacecolor</code>, <code>markeredgewidth</code>, <code>markeredgecolor</code>

```python
fig, ax = plt.subplots()

ax.plot(x, x + 1, color = "blue", label = 'x + 1', marker = 'o', markersize=8, markerfacecolor="red")
ax.plot(x, x + 2, color="purple", label = 'x + 2', lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
ax.legend()
```
