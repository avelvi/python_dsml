## Figure size, aspect ratio, DPI, saving figures

### Figure size, aspect ratio, DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the <code>figsize</code> and <code>dpi</code> keyword arguments

<li><code>figsize</code> is a tuple of the width and height of the figure in inches</li>
<li><code>dpi</code> is the dots-per-inch (pixel per inch).</li>


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure(figsize=(3, 2))

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(8, 2))

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(8, 2), dpi=100)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

The same we can do for subplots

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(8, 2))

axes[0].plot(x, y)
axes[1].plot(y, x)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(8, 2))

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```
