### Plot Types

There are several plot types built-in to pandas, most of them statistical plots by nature:

<ul>
    <li>df.plot.area</li>
    <li>df.plot.barh</li>
    <li>df.plot.density</li>
    <li>df.plot.hist</li>
    <li>df.plot.line</li>
    <li>df.plot.scatter</li>
    <li>df.plot.bar</li>
    <li>df.plot.box</li>
    <li>df.plot.hexbin</li>
    <li>df.plot.kde</li>
    <li>df.plot.pie</li>
</ul>

You can also just call <code>df.plot(kind='hist')</code> or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)

### Histograms

```python
plt.style.use('ggplot')
df1['A'].plot(kind = 'hist')
```

```python
df1['A'].plot.hist()
```

```python
df1['A'].plot.hist(bins = 50)
```

### Line Plots

```python
df1.head()
```

```python
df1.plot.line(x = df1.index, y = 'B')
```

```python
df1.plot.line(x = df1.index, y = 'B', figsize = (12,3), lw = 1)
```

### Scatter Plots

```python
df1.plot.scatter(x = 'A', y = 'B')
```

We can use <code>c</code> to <code>color</code> based off another column value:

```python
df1.plot.scatter( x = 'A', y = 'B', c = 'C')
```

 Use <code>cmap</code> to indicate colormap to use: 

```python
df1.plot.scatter(x = 'A', y = 'B', c = 'C', cmap = 'coolwarm')
```

Or use <code>s</code> to indicate size based off another column. 
<code>s</code> parameter needs to be an array, not just the name of a column:

```python
df1.plot.scatter(x = 'A', y = 'B', s = df1['C'])
```

But it can be scaled:

```python
df1.plot.scatter(x = 'A', y = 'B', s = df1['C'] * 50)
```

### BoxPlots

```python
df2.plot.box()
```

### Hexagonal Bin Plot

Useful for Bivariate Data, alternative to <code>scatterplot</code>:

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns = ['a', 'b'])
df.head()
```

```python
df.plot.hexbin(x = 'a', y = 'b')
```

```python
df.plot.hexbin(x = 'a', y = 'b', gridsize = 25)
```

```python
df.plot.hexbin(x = 'a', y = 'b', gridsize = 25, cmap = 'Oranges')
```
