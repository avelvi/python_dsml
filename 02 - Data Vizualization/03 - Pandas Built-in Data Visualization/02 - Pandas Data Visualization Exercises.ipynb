## Pandas Data Visualization Exercises

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df3 = pd.read_csv('df3')
df3.info()
```

```python
df3.head()
```

```python
df3.plot.scatter(x = 'a', y = 'b',
                 s = 50, figsize = (12, 3))
```

```python
df3.plot.scatter(x = 'a', y = 'b',
                 s = 50, c = 'red', figsize = (12, 3))
```

```python
df3['a'].plot.hist()
```

```python
plt.style.use('ggplot')
df3['a'].plot.hist()
```

```python
df3['a'].plot.hist(alpha = 0.5, bins = 25)
```

```python
df3['a'].plot.box()
```

```python
df3['b'].plot.box()
```

```python
df3[['a','b']].plot.box()
```

```python
df3['d'].plot.kde()
```

```python
df3['d'].plot.density(lw = 3, ls = '--')
```

```python
df3.plot.area(alpha = 0.4)
```

```python
df3.iloc[0:30].plot.area(alpha = 0.4)
```

```python
df3.iloc[0:30].plot.area(alpha = 0.4)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```
